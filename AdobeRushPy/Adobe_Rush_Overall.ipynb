```python
adobe = adobe[adobe["Language"]=="English"]
```

```python
#Lowercase all the text
adobe["Message"] = adobe["Message"].str.lower()
```

```python
#Remove unnneccessary punctuation
adobe["Message"] = adobe["Message"].str.replace('[^\w\s]','')
```

```python
adobe["Message"] = adobe["Message"].str.replace('/(\r\n)+|\r+|\n+|\t+/', " ")
```

```python
import nltk
nltk.download('stopwords')
```

```python
#Convert message type to string format
adobe['Message'] = adobe['Message'].astype(str)
```

```python
#Removing stopwords from text
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

def remove_stopwords(text):
    textArr = text.split(' ')
    rem_text = " ".join([i for i in textArr if i not in stop_words])
    return rem_text
```

```python
adobe['Message'] = adobe['Message'].apply(remove_stopwords)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#can only appear 95% but at least appear 2 times
cv = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
dtm = cv.fit_transform(adobe['Message'])
```

```python
#Get 4 topics
from sklearn.decomposition import LatentDirichletAllocation
LDA = LatentDirichletAllocation(n_components=4,random_state=42) 
LDA.fit(dtm)
```

```python
for index,topic in enumerate(LDA.components_):
    print(f'THE TOP 15 WORDS FOR TOPIC #{index}') #Grab last 15words
    print([cv.get_feature_names()[index] for index in topic.argsort()[-15:]])
    print('\n')
```

```python
topic_results = LDA.transform(dtm)
```

```python
adobe['Topic'] = topic_results.argmax(axis=1)
```

```python
def q1(score):
  if score <= 0.25: 
    return 1
  else: #successful
    return 0
```

```python
def q2(score):
  if score >= 0.25 and score < 0.5 : 
    return 1
  else: #successful
    return 0
```

```python
def q3(score):
  if score >= 0.5 and score < 0.75: 
    return 1
  else: #successful
    return 0
```

```python
def q4(score):
  if score >= 0.75: 
    return 1
  else: #successful
    return 0
```
